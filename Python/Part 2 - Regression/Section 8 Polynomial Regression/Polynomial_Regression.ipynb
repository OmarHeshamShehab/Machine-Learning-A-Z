```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Position_Salaries.csv")
X = dataset.iloc[:, 1:-1].values
y = dataset["Salary"]  # dataset.iloc[:, -1].values
```

Training Linear regression Model

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

Training Polynomial Regression Model

```python
from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures(degree=4) # this is the power of X
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

Visualizing Linear Regression Results

```python
plt.scatter(X, y, color="red")
plt.plot(X, lin_reg.predict(X), color="blue")
plt.title("Truth or bluff (Linear Regression)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

Visualizing Polynomial Regression Results

```python
plt.scatter(X, y, color="red")
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color="blue")
plt.title("Truth or bluff (Polynomial Regression)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

Visualising the Polynomoal Regression results (higher resolution and smoother curve)

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid),1)
plt.scatter(X,y,color='red')
plt.plot(X_grid,lin_reg_2.predict(poly_reg.fit_transform(X_grid)),color='blue')
plt.title("Truth or bluff (Polynomial Regression)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.legend()  # Show legend for different plots
plt.show()
```

Predicting new result with Linear Regression

```python
lin_reg.predict([[6.5]])
```

Predicting new result with Polynomial Regression

```python
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```
